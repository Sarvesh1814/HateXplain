```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ast import literal_eval
from scipy.stats import mode
```

```python
df = pd.read_json("dataset.json")
df=df.T
df=df.reset_index(drop=True)
df.head(5)
```

```python
df.shape
```

```python
# Getting Target class based on the majority as adviced in the paper 
target=[]
for i in range(len(df)):
    for j in range(3):
        
        if df.annotators[i][j]['target'] == []:
            pass
        else:
            x= df.annotators[i][j]['target'][0] 
            if x not in target:
                target.append(x)
                
Target = {}
for i in target:
    Target[i] = 0
```

```python
# Defining the label class and obtaining it for each data row 
df['Label_cat'] = 0
df["Target_cat"] = 0

for i in range(20148):
    Label={"normal":0,'hatespeech':0,'offensive':0}
    Targets = Target.copy()
    for j in range(3):
        if df.annotators[i][j]['target'] != []:
            if df.annotators[i][j]['label'] != []:
                Label[df.annotators[i][j]['label']]+=1
                Targets[df.annotators[i][j]['target'][0]]+=1
                label_cat = max(zip(Label.values(), Label.keys()))[1]        
                target_cat = max(zip(Targets.values(), Targets.keys()))[1]
    df.Target_cat[i]= target_cat
    df.Label_cat[i] = label_cat
```

```python
# Saving the processed dataset
dx = df.copy()
dx.drop(labels=['annotators'],axis='columns',inplace=True)
```

```python
#finding length of post_tokens
df['post_token_length'] = df['post_tokens'].apply(len)
df
```

```python
# Combining the rationales to get proper final rational to train our model effectively

# creating a dummy final rationale column to add the final converted rationale for each case
df['final_rationales']=""
for i in range(20148):
    #checking if the rationale for the column is empty or not
    if df['rationales'].iloc[i] != []:
        if len(df['rationales'].iloc[i]) ==2:
            if sum(df['rationales'].iloc[i][0]) >= sum(df['rationales'].iloc[i][1]): 
                df['final_rationales'].iloc[i] = df['rationales'].iloc[i][0] 
            else:
                df['final_rationales'].iloc[i] = df['rationales'].iloc[i][1] 
        else:
            df['final_rationales'].iloc[i]= list(mode(df['rationales'].iloc[i])[0][0])
    
    # filling rationales with 0 where the class is normal and traget is none
    else:
        df['final_rationales'].iloc[i]= [0] * df['post_token_length'].iloc[i]
```

```python
df=df[["post_id",'rationales','post_tokens','Target_cat','Label_cat','final_rationales']]
```
